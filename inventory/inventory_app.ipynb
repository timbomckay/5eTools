```python
# Firebase Setup:

cred_obj = firebase_admin.credentials.Certificate('../dnd-project-b2aa4-firebase-adminsdk-1ubid-87322187c7.json')

default_app = firebase_admin.initialize_app(cred_obj)
db = firestore.client()
```

```python
# Loads json data in and then sends it to Firebase:

with open("../data/items.json",'r') as f:
    data = json.loads(f.read())['item']

error_data = []

ct = 0
errnum = 1
for i in data:
    try:
        db.collection(u'Items').add(i)
        ct +=1
    except:
        print ("Error {}: Check value {} in data.".format(errnum, ct))
        ct += 1
        errnum += 1
        error_data.append(i)
        continue
```
