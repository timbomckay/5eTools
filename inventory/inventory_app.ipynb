```python
import json
import pandas as pd
```

```python
i = open("../data/items.json")
item_list = json.load(i)['item']
```

```python
# first off, get list of all item names; this will populate search bar?
item_names = []
for i in item_list:
    item_names.append(i['name'])
```

```python
def assign_item(ls, item):
    if item in item_names:
        ls.append(item)
    return ls
```
